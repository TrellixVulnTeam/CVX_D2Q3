```python
from IPython import embed
from skimage.feature import hog
from skimage import io
```

```python
%matplotlib inline
```

```python
im = io.imread('images/pikachu.jpg', as_gray=True)
normalised_blocks, hog_image = hog(im, orientations=9, pixels_per_cell=(
    8, 8), cells_per_block=(3, 3), visualise=True, transform_sqrt=True)
io.imshow(hog_image);
```
